```python
import cinrad
import warnings
warnings.filterwarnings("ignore")
```

```python
your_radar_file1 = (
    "/Users/lyp/Documents/cinrad/bz2/Z_RADR_I_Z9314_20200625140000_O_DOR_CB_CAP.bin"
)
your_radar_file2 = (
    "/Users/lyp/Documents/cinrad/bz2/Z_RADR_I_Z9317_20200625140000_O_DOR_SA_CAP.bin"
)

f1 = cinrad.io.CinradReader(your_radar_file1)
f2 = cinrad.io.CinradReader(your_radar_file2)
rls1 = list(f1.iter_tilt(230, "REF"))
rls2 = list(f2.iter_tilt(230, "REF"))
```

```python
gm = cinrad.calc.GridMapper([rls1[1],rls2[1]], max_dist=0.02)
grids = gm(step=0.005)
grids
```

```python
fig = cinrad.visualize.PPI(grids, add_city_names=True)
```
